# **Pseudoalgoritmos**

Un pseudoalgoritmo es una forma generalizada de implementar un algoritmo sin necesidad de incluir la sintaxis de un lenguaje de programación en concreto. Se caracteriza por no ser realmente ejecutable sino una representación de la idea lógica que implica su funcionamiento para que pueda ser transmitida arbitrariamentre. 

Dado un pseudoalgoritmo, el programador es libre de llevar a cabo esta idea lógica en cualquier lenguaje de programación y ejecutarla de una manera personalizada. Un ejemplo de pseudoalgoritmo sería el siguiente:

Este pseudoalgoritmo generaliza la idea del famoso algoritmo *Insertion sort*, que organiza en orden ascendente los números contenidos en un arreglo. Para ello inicia describiendo la información de entrada y la que se obtendrá como resultado, es decir el arreglo desordenado y el arreglo ordenado respectivamente. Luego, muestra el proceso mediante el cual se opera el arreglo para ordenarse, usando estructuras iterativas.

En particular, podemos implementar este pseudoalgoritmo en Python de la siguiente forma:

```python
# El algoritmo implementado es el insertion sort.
# Este ordena la lista de la siguiente manera:

def InsertionSort(A):
  N = len(A)  # Obtiene el tamaño de la lista.
  
  # Comienza desde el segundo elemento, tomando cada elemento
  # (key) y comparándolo con los anteriores.
  for j in range(2, N):
    key = A[j]  # Elemento a insertar.
    i = j - 1  # Índice del elemento anterior.
    
    # Desplaza los elementos mayores que 'key' hacia la derecha.
    while (i >= 0) and (A[i] > key):
      A[i + 1] = A[i]
      i = i - 1
    
    # Coloca 'key' en su lugar correcto.
    A[i + 1] = key
  
  return A  # Retorna la lista ordenada.


A = [1, 9, 2, 7, 1, 6, 0, 4]  # Lista desordenada.
B = InsertionSort(A)  # Ordena la lista.
print(B)  # Imprime la lista ordenada.
```

Los pseudoalgoritmos son fundamentales para transmitir ideas algorítmicas sin necesidad de restringirlas a un lenguaje de programación en concreto. Interpretación: El algoritmo de Euclides recibe dos números enteros a y b y busca su máximo común divisor (MCD) mediantes restas sucesivas. Primero, entra en un bucle mientras a y b sean diferentes. Dentro del bucle, si a es mayor que b, se actualiza a restándole b; de lo contrario, si b es mayor, se actualiza b restándole a. Este proceso se repite, reduciendo progresivamente los valores sin alterar el MCD, hasta que a y b sean iguales. Cuando esto ocurre, el algoritmo finaliza y devuelve ese valor como el máximo común divisor (MCD), ya que en ese punto ambos números han convergido al divisor común más grande.
------------------------------ -

```python
#----------------------------------------------------
#Ejecución: implemente y ejecute su algoritmo aquí
#----------------------------------------------------

# Definimos la función EuclideanAlgorithm que calcula 
# el Máximo Común Divisor (MCD) de dos números enteros a y b
def EuclideanAlgorithm(a, b):
    # Mientras los valores de 'a' y 'b' sean diferentes, seguimos iterando.
    while a != b:
        # Si 'a' es mayor que 'b', restamos 'b' a 'a'.
        if a > b:
            a = a - b
        # Si 'b' es mayor que 'a', restamos 'a' a 'b'.
        else:
            b = b - a
    # Cuando 'a' y 'b' sean iguales, ese valor es el MCD y lo retornamos.
    return a

print(EuclideanAlgorithm(10, 5))
```

 El objetivo de este algoritmo es simular el cambio continuo de una variable, actualizando el valor de la posición en el paso siguiente $x(k+1)$ a partir del valor de la posición en el paso actual $(x(k))$ y la forma en que cambia $d(k)$ durante $n$ pasos. 
 
 Este es el tipo de algoritmos usados en inteligencia artificial para guiar los procesos de aprendizaje de las máquinas. En este ejemplo, tenemos un proceso iterativo de $n$  pasos que actualiza un arreglo $x$. El algoritmo recibe como parámetros de entrada una condición inicial $x[0]$ y el número de pasos $n$, y debe retornar el arreglo final $x$, que muestra los sucesivos cambios de la variable.

**2.1.** Implemente el algoritmo anterior como una función llamada ***trajectory*** que recibe como parámetros una condición inicial $x_0$ de tipo float y un número de pasos ***n*** de tipo int y realice una prueba de ejemplo.  

```python
#----------------------------------------------------
# Ejecución: implemente y ejecute su algoritmo aquí
#----------------------------------------------------

# Definimos la función 'trajectory' que calcula una secuencia de valores en base a un proceso iterativo.
# Parámetros:
# - 'valor': el valor inicial desde el cual comienza la iteración.
# - 'steps': la cantidad de pasos a realizar en la iteración.
def trajectory(valor, steps):
    # Creamos una lista 'array_x' que almacena los valores calculados, 
    # comenzando con el valor inicial proporcionado.
    array_x = [valor]
    
    # Creamos una lista 'array_d' para almacenar los valores intermedios derivados de la función coseno.
    array_d = []
    
    # Iteramos desde 0 hasta 'steps - 1' para generar la trayectoria de valores.
    for i in range(0, steps - 1):
        # Calculamos un valor basado en la función coseno y lo almacenamos en 'array_d'.
        array_d.append(np.cos(array_x[i]) + (10/3) * np.cos((10/3) * array_x[i]))
        
        # Actualizamos 'array_x' restando un pequeño ajuste (0.1 * array_d[i]) al último valor.
        array_x.append(array_x[i] - 0.1 * array_d[i])
    
    # Retornamos la lista con todos los valores calculados en 'array_x'.
    return array_x

# Llamamos a la función con un valor inicial de 8 y realizamos 5 iteraciones
print(trajectory(8, 5))
```
